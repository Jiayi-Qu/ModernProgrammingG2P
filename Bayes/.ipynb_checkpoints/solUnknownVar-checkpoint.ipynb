```julia
using Distributions
using StatsBase
using Printf
using LinearAlgebra
using ProgressMeter
```


1 . Generate data from model (1) with  ùëò=10  covariates.

```julia
n = 2000 #number of observations
k = 10   #number of covariates
M = rand(Binomial(2,0.5),n,k)
M = M .- mean(M,dims=1)
X = [ones(n) M]
```

```julia
betaTrue = [1; randn(k)]
y = X*betaTrue+ randn(n);
```

### Exercise: 

1. Extend the sampler to treat  $ùúé^2_ùõΩ$  as an unknown. Plot the posterior distribution for  $ùúé^2_ùõΩ$ .

```julia
function GibbsSample2(y,X,b,niter,ŒΩB,Sb)
    chain = zeros(niter)   #to save sigma2
    n,k = size(X)
    Œª    = 1.0
    @showprogress "drawing MCMC samples..." for iter = 1:niter
        # sampling intercept
        w = y - X[:,2:end] * b[2:end]
        x = X[:,1]
        xpxi = 1/(x'x)[1,1]
        bHat = (xpxi*x'w)[1,1] 
        b[1] = rand(Normal(bHat, sqrt(xpxi))) # using residual var = 1 
        # sampling marker effects
        for i=2:k
            btemp = copy(b)
            btemp[i] = 0
            w = y - X*btemp
            x = X[:,i]
            xpxi = 1/((x'x) + Œª)[1,1]
            bHat = (xpxi*x'w)[1,1]
            b[i] = rand(Normal(bHat, sqrt(xpxi))) # using residual var = 1  
        end         
        # sampling sigmaB
        varb = (sum(b[2:end].^2) + ŒΩB*Sb) /rand(Chisq(ŒΩB+k-1))
        Œª    = 1.0/varb
        chain[iter] = varb
    end
    return chain
end
```

```julia
niter = 1000                     # number of samples
b     = zeros(k+1)
ŒΩB    = 4
varb  = 1
vare  = 1
Sb    = varb*(ŒΩB - 2)/ŒΩB
@time chain = GibbsSample2(y,X,b,niter,ŒΩB,Sb);
```

```julia
using Plots
```
